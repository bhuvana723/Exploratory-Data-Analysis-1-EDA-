```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/Cardiotocographic.csv")
data
```

```python
data.info()
```

```python
data.describe()
```

```python
missing_values = data.isnull().sum()
data.fillna(data.mean(), inplace=True)


```

```python
data.info()
```

```python
Q1 = data['Width'].quantile(0.25)
Q3 = data['Width'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_filtered = data[(data['Width'] >= lower_bound) & (data['Width'] <= upper_bound)]
data['Width'] = np.clip(data['Width'], lower_bound, upper_bound)
```

```python
Q1 = data['ALTV'].quantile(0.25)
Q3 = data['ALTV'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_filtered = data[(data['ALTV'] >= lower_bound) & (data['ALTV'] <= upper_bound)]
data['ALTV'] = np.clip(data['ALTV'], lower_bound, upper_bound)
```

```python
Q1 = data['ASTV'].quantile(0.25)
Q3 = data['ASTV'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_filtered = data[(data['ASTV'] >= lower_bound) & (data['ASTV'] <= upper_bound)]
data['ASTV'] = np.clip(data['ASTV'], lower_bound, upper_bound)
```

```python
dQ1 = data['LB'].quantile(0.25)
Q3 = data['LB'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
data_filtered = data[(data['LB'] >= lower_bound) & (data['LB'] <= upper_bound)]
data['LB'] = np.clip(data['LB'], lower_bound, upper_bound)
```

```python
data.describe()
```

```python
for col in data.select_dtypes(include=['number']):
  Q1 = data[col].quantile(0.25)
  Q3 = data[col].quantile(0.75)
  IQR = Q3 - Q1
  print(f"IQR for {col}: {IQR}")
```
